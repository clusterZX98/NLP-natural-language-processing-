# Regex Library

```python
import re
text = "Contact us at support@gmail.com or call 9876543210"
print("Given Text : " , text)
email = re.findall(r'\S+@\S+', text)
phone = re.findall(r'\d{10}', text)

print("Email : " , email)
print("Phone No. : " , phone)
```

# Spacy Library

```python
import spacy
nlp = spacy.load("en_core_web_sm")
text = "Apple opens a new office in california"
print("Given Text : " , text)
doc = nlp(text)
for i in doc.ents:
  print(i.text , i.label_)
```

# KerasNLP

```python
import keras_nlp
preprocessor = keras_nlp.models.BertPreprocessor.from_preset(
"bert_tiny_en_uncased"
)
classifier = keras_nlp.models.BertClassifier.from_preset(
"bert_tiny_en_uncased",
num_classes=5,
preprocessor=preprocessor
)

prediction = classifier.predict(
[
  "Artificial Intelligence is transforming industries across the world by enabling machines to learn from data and make intelligent decisions. From healthcare and finance to education and transportation, AI-powered solutions are helping organizations improve efficiency, reduce costs, and enhance customer experiences. Recent advancements in generative AI, large language models, and natural language processing have made it possible for businesses to automate complex tasks, generate high-quality content, and gain valuable insights from unstructured data. As AI adoption continues to grow, professionals who develop skills in machine learning, data science, and AI engineering will have access to numerous career opportunities in the future."

]
)
print(prediction)
```

# Stanza Library

```python
!pip install stanza
```

```python
import stanza
stanza.download('en')
nlp = stanza.Pipeline('en')
doc = nlp("Natural Language Processing is powerful.")
for sentence in doc.sentences:
  for word in sentence.words:
    print(word.text, word.upos)
```

# FastText

```python
from gensim.models import FastText
sentences = [
["I", "love", "NLP"],
["FastText", "is", "efficient"]
]
model = FastText(
sentences,
vector_size=10,
min_count=1
)
print(model.wv['NLP'])
```
