```python
import nltk
nltk.download('all')
```

```python
# tokenization using nltk
from nltk import word_tokenize,sent_tokenize
para = "hello, i'm piyush.i am studying nlp."
print("original sentence",para)
print(word_tokenize(para))
print(sent_tokenize(para))
```

```python
# STOPWORDS REMOVAL that creates noises

nltk.download('stopwords')
from nltk.corpus import stopwords

text = "I am learning NLP in the classroom"
stop_words = set(stopwords.words('english'))
filter = [i for i in text if i.lower() not in stop_words]
print("text after removing stop words:",filter,"\n")
```

```python
# STOPWORDS REMOVAL that creates noises

nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk import word_tokenize
text = "I am learning NLP in the classroom"
tokens = word_tokenize(text)
stop_words = set(stopwords.words('english'))
filter = [i for i in tokens if i.lower() not in stop_words]
print("text after removing stop words:",filter,"\n")
```

```python
#stemming & lemmatization
from nltk.stem import PorterStemmer
port = PorterStemmer()

print(port.stem("going"))
print(port.stem("went"))
print(port.stem("goes"))

words = ["playing", "played", "player","miscommunication"]
print("original list:",words)
for i in words:
  print(i,":",port.stem(i))
```

```python
# Lemmatization

nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
word = ("playing","goose","mice","goes","ran","tooth","better","drove")
print("original words:",word)

for i in word:
  print(i, "->", lemma.lemmatize(i))
```

```python
# POS(part of speech) Tagging

nltk.download('averaged_perceptron_tagger')
senten = "the dog run faster than jai"
tokens = nltk.word_tokenize(senten)
print(tokens)
pos_tag = nltk.pos_tag(tokens)
print(pos_tag)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
text = ("Elon Musk issued the share of SpaceX and become the First Trillionaire")
print("original text:",text)
doc = nlp(text)
for i in doc.ents:
  print(i.text,i.label_)
```
